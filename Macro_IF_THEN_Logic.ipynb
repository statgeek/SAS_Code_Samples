This workbook is an example of how to create a macro to execute if then logic. 

```sas
%macro formatter(file_type, file, start_date);

%put This is my macro;

%mend;

%formatter(None, none, none);
```

```sas
/*Lets add the logic for the different file_types*/
%macro formatter(file_type, file, start_date);

%if &file_type=1 %then %do;
    %put file_type = &file_type. ;
%end;

%else %if &file_type=2 %then %do;
    %put file_type = &file_type.;
%end;

%else %if &file_type=3 %then %do;
    %put file_type = &file_type.;
%end;

%else %do;
    %put Invalid Entry;
%end;

%mend;

%formatter(1, none, none);
%formatter(2, none, none);
%formatter(3, none, none);
%formatter(4, none, none);
```

```sas
/*Switch from test to full macro call*/
%macro formatter(file_type, file, start_date);

%if &file_type=1 %then %do;
    %converter1(&file, &start_date) ;
%end;

%else %if &file_type=2 %then %do;
    %converter1(&file, &start_date);
%end;

%else %if &file_type=3 %then %do;
    %converter1(&file, &start_date);
%end;

%else %do;
    %put Invalid Entry;
%end;

%*Any other code goes afterwards;

%mend;

%formatter(1, none, none);
```
